```python
import pandas as pd
```

```python
df = pd.read_csv('global_superstore_2016.csv',encoding = 'latin-1')
df
```

```python
df["Sales"] = df["Sales"].str.replace(r"[$,()]", "", regex = True)
df["Profit"] = df["Profit"].str.replace(r"[$,()]", "", regex = True)
df[["Order Date","Ship Date"]] = df[["Order Date","Ship Date"]].replace("-","/",regex = True)
df
```

```python
# change data type 
df["Sales"] = df["Sales"].astype(float)
df["Profit"] = df["Profit"].astype(float)
df.info()
```

```python
# change data type 
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
df.info()
```

```python
# 3. Get the list of Order ID's where the Indian customer's have bought the things under the category 'Technology' after paying the Shipping Cost more than 500.
Indian_cust = df[(df['Country'] == 'India') & (df['Category'] == 'Technology') & (df['Shipping Cost'] >500)]
Indian_cust[['Order ID','Country','Category','Shipping Cost']]
```

```python
# 4. Get the list of Order ID's where the Indian customer's have bought the things under the category 'Technology' the Sales greater than 500.
Indian_sales = df[(df["Country"] == "India") & (df["Category"] == "Technology") & (df["Sales"] > 500)]
Indian_sales[["Order ID","Country","Category","Sales"]]
```

```python
# 5. How many people from the State 'Karnataka' have bought the things under the category 'Technology'?
karnatka_people = df[(df["State"] == "Karnataka") & (df["Category"] == "Technology")]
karnatka_people[['Customer Name']].count()
```

```python
# 6. Get the list of countries where the 'Profit' and 'Shipping Cost's are greater than or equal to 2000 and 300 respectively.
countries_list = df[(df["Profit"] >= 2000) & (df["Shipping Cost"] >= 300)]
countries_list[["Country", "Profit", "Shipping Cost"]].sort_values("Country")
```

```python
# 7. Find the list of Indian states where the people have purchased the things under the category Technology.
tech_product = df[(df["Country"] == "India") & (df["Category"] == "Technology")]
tech_product["State"]
```

```python
# 8. Find the overall rank of "India" where the 'Profit' is maximum under the category 'Technology'.
Rank = df[df["Category"] == "Technology"]
Rank[["Country","Profit"]].groupby("Country").sum().sort_values("Profit",ascending = False)
```
